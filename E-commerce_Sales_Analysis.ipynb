```python
top_customers= df.groupby('Customer Name')['Order ID'].count().sort_values(ascending= False).head(10)
top_customers
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_customers.values, y=top_customers.index, hue=top_customers.index, palette='Blues_r', legend= False)
plt.title('Top 10 Customers by Orders')
plt.xlabel('Number of Orders')
plt.ylabel('Customer Name')
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
top_regions= df.groupby('Region')['Order ID'].count().sort_values(ascending= False)
top_regions
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_regions.values, y=top_regions.index, hue=top_regions.index, palette= 'coolwarm', legend= False)
plt.title('Orders by Region')
plt.xlabel('Number of Orders')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

```python
top_category= df.groupby('Category')['Order ID'].count().sort_values(ascending=False)
top_category
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_category.values, y=top_category.index, hue=top_category.index, palette= 'viridis', legend= False)
plt.title('Orders by Category')
plt.xlabel('Number of Order')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Discount', y='Sales', alpha=0.5, color='steelblue')
plt.title('Effects of Discount on Sales')
plt.xlabel('Discount %')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
category_sales_profit= df.groupby('Category')[['Sales','Profit']].sum().sort_values('Sales',ascending=False)
category_sales_profit.plot(kind='barh', figsize=(10,6), color=['steelblue','coral'])
plt.title('Sales and Profit by Category')
plt.xlabel('Amount')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
df['Month']= df['Order Date'].dt.to_period('M')
monthly_sales= df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot(kind= 'line', color='steelblue', marker='o')
plt.title('Monthly Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
payment_counts= df['Payment Mode'].value_counts()
payment_counts
```

```python
plt.figure(figsize=(8,8))
plt.pie(payment_counts.values, labels= payment_counts.index, autopct='%1.1f%%', startangle=140, colors=['steelblue', 'coral', 'green', 'gold', 'purple'])
plt.title('Payment Mode Distribution')
plt.tight_layout()
plt.show()
```
